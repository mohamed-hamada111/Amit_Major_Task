### Department Class

#### Used to define each department in a specific hospital, manage the number of employees and the number of patients in each department.




```python
import import_ipynb

import patient_class
from patient_class import Patient
from Staff import Staff
```

```python
class Department:
    '''define each department in the hospital and manage the number of patient and staff members in each department'''

    def __init__(self,name):
        self.name = name
        self.patients =[]
        self.staff =[]
    
    def add_patient(self, patient : Patient):
        '''Function used to manage the patients and add them for certain Department
        
        :param patient: the patient will be added to the department.
        :return None:'''
        self.patients.append(patient)
        print(f'Patient {patient.name} added to {self.name} department')

    
    def add_staff(self, staff_member : Staff):
        '''Function used to manage the staff members and add them for certain Department
        
        :param staff_member: the member will be added to the department.
        :return None:'''
        self.staff.append(staff_member)
        print(f'Staff {staff_member.name} added to {self.name} department')

    def display_info(self):
        '''Function used to show all information related to the Department.'''
        print(f'The Department name is {self.name}')
        print(f'The number of patients : {len(self.patients)}')
        print(f'The number of staff members : {len(self.staff)}')


        

    
```

```python
dep = Department('Cardiology')

patient1 = Patient("Ahmed", 30, "The patient has no allergies and a recorded history of flu")
staff1 = Staff('mohamed' ,21 ,'doctor')
staff2 = Staff('khaled' ,22 ,'doctor')


dep.add_patient(patient1)
dep.add_staff(staff1)
dep.add_staff(staff2)
dep.display_info()
```
